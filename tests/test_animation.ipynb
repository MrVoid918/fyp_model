```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.animation as animation
import matplotlib.patches as patches
from matplotlib.transforms import Bbox
from bounding_box_plotter import BoundingBoxPlotter
import cv2

import numpy as np

%matplotlib widget
```

```python
fig, ax = plt.subplots()
```

```python
ims = []

bbplotter = BoundingBoxPlotter()
data = np.loadtxt("pred.txt", delimiter=',')
data = np.loadtxt("output/20230315_4_60_pred.txt", delimiter=',')

for i in data[:, 0]:
    from pathlib import Path
    path = Path(f"inference/images/img1/{str(int(i)).zfill(6)}.jpg").resolve()
    
    im = cv2.imread(str(path))
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

    bbplotter.draw_mot(im, data[data[:, 0] == i])
    
    im = ax.imshow(im, animated=True)
    ims.append([im])

    plt.draw()
```

```python
ani = animation.ArtistAnimation(fig, ims, blit=True, repeat_delay=100)
```

```python
plt.show()
```
